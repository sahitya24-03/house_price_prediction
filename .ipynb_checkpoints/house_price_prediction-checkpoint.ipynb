```python
numeric_df=df.select_dtypes(include=[np.number]) #only numeric features
corr= numeric_df.corr() #correlation matrix

plt.figure(figsize=(16,10))
sns.heatmap(corr,cmap='coolwarm',annot=False)
plt.title("correlation hetamap")
plt.show()
```

### Correlation with SalePrice (Top Features)

* We selected only numeric columns with select_dtypes to avoid mistakes with text data. Then we calculated the correlation of each numeric feature with SalePrice and ordered them in descending order.

* This enables us to identify the determining characteristics shaping housing costs.ing.


```python
plt.figure(figsize=(10,6))
sns.boxplot(x='OverallQual',y='SalePrice', data=df)
plt.xlabel('overall quantity(1=very poor,10=good)')
plt.ylabel("sale price(USD)")
plt.show()
```

### Sale Price Compared to Overall Quality

The box plot exhibits a positive relationship between `SalePrice` and `OverallQual`.

* The finish and material quality improve (from 1 to 10) and the cost also increases.
* Most are situated in quality levels 5 through 7.
* Better quality homes (8–10) sell with significantly higher prices.

This confirms that `OverallQual` is a significant attribute for the prediction of price.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df, color='blue')
plt.title("SalePrice vs Above Ground Living Area")
plt.xlabel("GrLivArea (Sq ft)")
plt.ylabel("SalePrice (USD)")
plt.show()
```

### SalePrice vs GrLivArea

A positive correlation between `GrLivArea` (above-ground living area, measured in square feet) and `SalePrice` is evident from the scatter plot.

- The larger the living space, the higher the cost of the house.  
- Some large homes are priced significantly lower than expected — these are **outliers**.

This confirms that `GrLivArea` is an important factor for our model.
