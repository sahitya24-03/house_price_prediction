```python
#droping high missing values
df.drop(['Id', 'PoolQC', 'Alley', 'Fence', 'MiscFeature'], axis=1, inplace=True)
```

```python
# Numeric columns → Fill with median
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Categorical columns → Fill with mode
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode())
```

###  Handling Missing Values

We cleaned the dataset by:

-  Dropping columns with too many missing values: `PoolQC`, `Alley`, `Fence`, `MiscFeature`
-  Filling missing values in **numeric columns** with the column **median**
-  Filling missing values in **categorical columns** with the column **mode**

This ensures the dataset is complete and ready for encoding and model training.


```python
df_encoded = pd.get_dummies(df, drop_first=True)
```

###  Encoding Categorical Variables

We applied **One-Hot Encoding** using `pd.get_dummies()` to convert all categorical features into numerical format.

This is necessary as machine learning models cannot work with text-based columns.


```python
from sklearn.model_selection import train_test_split
#1.separate features and target 
x = df_encoded.drop('SalePrice',axis=1)
y = df_encoded['SalePrice']
```

```python
#2. split into train and test sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

###  Train-Test Split

We split the dataset into training and testing sets using `train_test_split()`:

- 80% for training
- 20% for testing
- `random_state=42` ensures reproducibility


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
```

```python
#1. Initialize and train the model
lr_model = LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
#2.make predictions
y_pred = lr_model.predict(x_test)
```

```python
#3.evaluate the model
mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
r2=r2_score(y_test,y_pred)
```

```python
print("Linear RegressionResults:")
print("MAE:",mae)
print("RMSE:",rmse)
print("R² score:",r2)
```

###  Linear Regression Results

- **MAE:** \$20,434  
- **RMSE:** \$52,298  
- **R² Score:** 0.64

This provides a good baseline. The model explains 64% of price variation, but there's room to improve with advanced models like Random Forest or Gradient Boosting.


```python
from sklearn.ensemble import RandomForestRegressor
#1. Initialize and train the model
rf_model=RandomForestRegressor(random_state=42)
rf_model.fit(x_train,y_train)
```

```python
#2.Make predictions
rf_pred = rf_model.predict(x_test)
```

```python
#3.Evaluate the model 
rf_mae= mean_absolute_error(y_test,rf_pred)
rf_rmse=np.sqrt(mean_squared_error(y_test,rf_pred))
rf_r2=r2_score(y_test,rf_pred)
```

```python
print("Random Forest Results:")
print("MAE:",rf_mae)
print("RMSE:",rf_rmse)
print("R² score:",rf_r2)
```

### Random Forest Results
- **MAE**: \$17,453
- **RMSE**: \$28,766
- **R²** score: 0.89

###  Model Comparison

| Metric      | Linear Regression | Random Forest |
|-------------|-------------------|----------------|

 **Conclusion:**

```python
from sklearn.ensemble import GradientBoostingRegressor
#1. Initialize and train the model
gb_model=GradientBoostingRegressor(random_state=42)
gb_model.fit(x_train,y_train)
```

```python
#2. make predictions
gb_pred=gb_model.predict(x_test)
```

```python
#3.evaluate the model
gb_mae=mean_absolute_error(y_test,gb_pred)
gb_rmse=np.sqrt(mean_squared_error(y_test,gb_pred))
gb_r2=r2_score(y_test,gb_pred)
```

```python
print("gradient boosting results")
print("MAE:",gb_mae)
print("RMSE:",gb_rmse)
print("R² Score:",gb_r2)
```

### Gradient boosting results
